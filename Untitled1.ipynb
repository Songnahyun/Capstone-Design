```python
import librosa
import pandas as pd
import librosa.display
import IPython.display
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
```

```python
audio_path = '뭔가 싫어서 우는.wav'
y, sr = librosa.load(audio_path) #y: 소리가 떨리는 세기를 순서대로 나열, sr: 1초에 n개의 샘플을 뽑음

print('Sampling rate (Hz): %d' %sr)
print('Audio length (seconds): %.2f' % (len(y) / sr)) #음악의 길이(초) = 음파의 길이/Sampling rate
```

```python
#FFT
fft = np.fft.fft(y) 

magnitude = np.abs(fft)
frequency = np.linspace(0, sr, len(magnitude))

#푸리에 변환을 통과한 specturm은 대칭구조로 나와서 high frequency 부분 절반을 날리고 앞쪽 절반만 사용한다.
left_frequency = frequency[:int(len(frequency)/2)]
left_magnitude = magnitude[:int(len(magnitude)/2)]

fig = plt.figure(figsize = (14,5))
plt.plot(left_frequency, left_magnitude)
```

```python
#주파수 영역에서 분석(데시벨 scaling을 적용한 log-spactogram)
X = librosa.stft(y) #시간영역의 값을 주파수 영역에서 표현(STFT,시간 정보가 유실되는 것을 방지하기 위해)

X_mag = abs(X) #절대값

# energy levels(dB) 로 변경
Xdb = librosa.amplitude_to_db(X_mag)

plt.figure(figsize=(20, 5))
librosa.display.specshow(Xdb, sr = 15000 , x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
#MFCC(특징값 추출)

mfcc = librosa.feature.mfcc(y=y, sr=sr)

fig = plt.figure(figsize = (14,5))
librosa.display.specshow(mfcc, 
                         sr=sr, 
                         x_axis='time',)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar(format='%+2.0f dB')
plt.show()
```

```python
#Mel spectogram
S = librosa.feature.melspectrogram(y, sr=sr)
S_DB = librosa.amplitude_to_db(S, ref=np.max)

plt.figure(figsize=(16,6))
librosa.display.specshow(S_DB, sr=sr, hop_length = 10, x_axis='time',y_axis='log')
plt.colorbar()
plt.show()
```
